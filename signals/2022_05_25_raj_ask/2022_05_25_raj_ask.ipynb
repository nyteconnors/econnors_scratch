```python
import matplotlib.pyplot as plt

def graph(plt, data, labels, bins, title): 
    plt.hist(data[0], bins = bins, alpha=0.5, label = labels[0])
    plt.hist(data[1], bins = bins, alpha=0.5, label = labels[1])
    plt.title(title)
    plt.legend(loc="upper right")   
    plt.savefig("plots/" + title + ".png", facecolor='white', transparent=False)
    plt.close()
```

```python
unique_users = data.drop_duplicates(["type", "regi_id", "subscriber_start_ts"])
january = unique_users[unique_users["type"] == "january"].tenure
october = unique_users[unique_users["type"] == "october"].tenure
graph(plt, [january, october], ["january", "october"], 100, "tenure")
```

```python
january = data[(data["month"] == 1) & (data["type"] == "january")].active_days
october = data[(data["month"] == 10) & (data["type"] == "october")].active_days
graph(plt, [january, october], ["january", "october"], 30, "active_days")
```
